What is a random forest
A random forest consists of multiple random decision trees. Two types of randomnesses are built into the trees. First, each tree is built on a random sample from the original data. Second, at each tree node, a subset of features are randomly selected to generate the best split. (https://towardsdatascience.com/random-forest-3a55c3aca46d#:~:text=Clustering%20with%20random%20forests,-Similar%20to%20outlier&text=The%20procedure%20is%20similar%20to,built%20for%20the%20classification%20problem.)

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=10000, centers=8,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

Changing the samples to 10000 and centers to 8. We have significantly increased the amount of data points and number of centers which has caused the red cluster to move further away from the rest. 

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=10000, centers=8,
                  random_state=3, cluster_std=5.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

I have changed random_state to 3 and cluster_std to 5.0. This has caused the clusters to spread out on the feature space. 

cluster_std: the parameter for this determines how spread out the clusters are, so the higher the value the more scattered each cluster will be
